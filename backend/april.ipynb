```python
import numpy as np
import librosa
import cv2
from tensorflow.keras.models import load_model
from sklearn.preprocessing import LabelEncoder
import os

# --- CONFIGURE THESE PATHS ---
MODEL_PATH = '/Users/rhea.dmello/Desktop/Code/Beehive Project/backend/queenbee_final_model.h5'
AUDIO_FILE_PATH = "/Users/rhea.dmello/Desktop/Code/Beehive Project/backend/No queen bee.mp3"

# Check if files exist
if not os.path.exists(MODEL_PATH):
 raise FileNotFoundError(f"Model not found at: {MODEL_PATH}")

if not os.path.exists(AUDIO_FILE_PATH):
 raise FileNotFoundError(f"Audio file not found: {AUDIO_FILE_PATH}")

# Load model
model = load_model(MODEL_PATH)
print("✅ Model loaded. Input shape:", model.input_shape)

# Re-create label encoder
encoder = LabelEncoder()
encoder.fit(['bee', 'nobee', 'noqueen'])

# Feature extraction — mel-spectrogram image (128x128x3)
def extract_features(file_path, target_sr=16000):
 try:
 audio_data, original_sr = librosa.load(file_path, sr=None)
 audio_data = librosa.resample(audio_data, orig_sr=original_sr, target_sr=target_sr)

 mel = librosa.feature.melspectrogram(y=audio_data, sr=target_sr, n_mels=128)
 mel_db = librosa.power_to_db(mel, ref=np.max)

 mel_img = cv2.resize(mel_db, (128, 128))
 mel_img = (mel_img - mel_img.min()) / (mel_img.max() - mel_img.min()) * 255
 mel_img = mel_img.astype(np.uint8)
 mel_img = np.stack([mel_img] * 3, axis=-1)

 return mel_img.reshape(1, 128, 128, 3)

 except Exception as e:
 print(f"❌ Error processing {file_path}: {e}")
 return None

# --- PREDICTION ---
print(f"\n🎧 Loading audio: {os.path.basename(AUDIO_FILE_PATH)}")
features = extract_features(AUDIO_FILE_PATH)

if features is not None:
 prediction = model.predict(features, verbose=0)
 predicted_class = encoder.inverse_transform([np.argmax(prediction)])​⟦CITE:0⟧​
 confidence = np.max(prediction) * 100

 print(f"\n✅ Predicted Class : {predicted_class.upper()}")
 print(f"📊 Confidence : {confidence:.2f}%")
else:
 print("❌ Failed to extract features from audio.")
```
